```python

path_to_data    = "./data"
path_to_output  = "./docs/DMI"
csv_args = {"sep":"|", 
            "skiprows":[0], 
            "index_col":"IRI", 
            "comment":'#'}

# Organisation types (SDO, ORG, EDU, COM, ...)
organisation_types = pd.read_csv(f"{path_to_data}/data.csv", **csv_args) 
for IRI_orgtype, row in organisation_types.iterrows():
    p_orgtype = Path(path_to_output, f"{IRI_orgtype}")
    p_orgtype.mkdir(parents=True, exist_ok=True)
    #TODO: fill orgtype template
    
    # Organisations
    organisations = pd.read_csv(f"{path_to_data}/{IRI_orgtype}/{IRI_orgtype}.csv", **csv_args)
    for IRI_org, row in organisations.iterrows():
        p_org = Path(p_orgtype, f"{IRI_org}")
        p_org.mkdir(parents=True, exist_ok=True)
        
        # Specs
        specifications = pd.read_csv(f"{path_to_data}/{IRI_orgtype}/{IRI_org}/{IRI_org}.csv", **csv_args)
        for IRI_spec, row in specifications.iterrows():
            p_spec = Path(p_org, f"{IRI_spec}")
            p_spec.mkdir(parents=True, exist_ok=True)
            
            # DMIs
            dmis = pd.read_csv(f"{path_to_data}/{IRI_orgtype}/{IRI_org}/{IRI_spec}/{IRI_spec}.csv", **csv_args)
            for IRI_dmi, row in dmis.iterrows():
                # Create files and folders in docs/... 
                suffixes  = [".ttl",           ".html",           ".xml"          ]
                templates = [ttl_dmi_template, html_dmi_template, xml_dmi_template]
                substitutes = dict()
                substitutes["IRI_orgtype"]            = IRI_orgtype
                substitutes["IRI_org"]                = IRI_org
                substitutes["IRI_spec"]               = IRI_spec
                substitutes["IRI_dmi"]                = IRI_dmi
                substitutes["Literal_skos_prefLabel"] = row["skos:prefLabel"]
                substitutes["Literal_skos_altLabel"]  = row["skos:altLabel"]
                for suffix, template in zip(suffixes, templates):
                    p_dmi = Path(p_spec, f"{IRI_dmi}{suffix}")
                    with p_dmi.open(mode="w", encoding="UTF8") as fh:
                        fh.write(template.substitute(substitutes))
            
    
    
```
